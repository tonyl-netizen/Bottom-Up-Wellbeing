```python
import os
from openai import OpenAI
import time

here = '1_Method/'
```

```python
start_index = 0
end_index = 100
```

```python
in_file = here + 'data/ideal_life_adjectives.pkl'
out_file = here + f'data/DGP/adj_{start_index}_{end_index}.csv'

adjectives = ['beautiful', 'happy']

adjectives = adjectives[start_index:end_index]
```

```python
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

prompt = (here / "data" / "ideal_life_prompt.txt").read_text(encoding="utf-8")
system_template = (here / "prompts" / "system_message.txt").read_text(encoding="utf-8")

def assess_comment(adjective: str):
    system_message = system_template.format(adjective=adjective)
    try:
        response = client.responses.create(
            model="gpt-4o-mini",
            input=[
                {"role": "system", "content": system_message},
                {"role": "user", "content": prompt},
            ],
        )
        return response.output_text
    except Exception as e:
        print(f"Error processing request for adjective '{adjective}': {e}")
        return None
```

```python
from tqdm import tqdm
import pandas as pd
import time

rows = []

for adj in tqdm(adjectives):
    for i in range(2):
        resp = assess_comment(adj)
        rows.append({"adj": adj, "response": resp})
        time.sleep(1)

output_df = pd.DataFrame(rows)
output_df.to_csv(out_file, index=False)
```
